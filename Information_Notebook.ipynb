## Pivot Table of Shipments per month and industry.

```python
mean_values = df.pivot_table(values='shipments_per_month', index='industry', aggfunc=['mean', 'median'])    
print(mean_values)
```

```python
pivot_table_a1 = df.pivot_table(values=['monthly_spend_usd','avg_delay_hours'], index='industry')    
print(pivot_table_a1)
```
